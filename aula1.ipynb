```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
img_path = list(uploaded.keys())[0]
img = cv2.imread(img_path)


print(f'Dimenssões: {img.shape}')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('Imagem RBG')
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(f'Dimenssões: {img_gray.shape}')
plt.imshow(img_gray, cmap='gray')
plt.title('Imagem gray')
plt.axis('off')
plt.show()
```

```python
hist = cv2.calcHist([img_gray], [0], None, [256], [0,256])
plt.plot(hist)
plt.title('Histograma da Imagem em Escala de Cinza')
plt.xlabel('Nivel de Intensidade')
plt.ylabel('Numero de Pixels')
plt.show()
```
