```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
img_path = list(uploaded.keys())[0]
img = cv2.imread(img_path)


print(f'Dimenss√µes: {img.shape}')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('Imagem RBG')
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(f'Dimenss√µes: {img_gray.shape}')
plt.imshow(img_gray, cmap='gray')
plt.title('Imagem gray')
plt.axis('off')
plt.show()
```

```python
hist = cv2.calcHist([img_gray], [0], None, [256], [0,256])
plt.plot(hist)
plt.title('Histograma da Imagem em Escala de Cinza')
plt.xlabel('Nivel de Intensidade')
plt.ylabel('Numero de Pixels')
plt.show()
```

```python
edges = cv2.Canny(img_gray, 100, 200)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(edges, cmap='gray')
plt.title('Segmenta√ß√£o por Bordas')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(img_gray, None)
img_sift = cv2.drawKeypoints(img_gray, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(10,5))
plt.imshow(img_sift, cmap='gray')
plt.title('Pontos Chaves Detectados')
plt.axis('off')
plt.show()
```
